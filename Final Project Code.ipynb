```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data Preparation

```python
# load the csv data

df = pd.read_csv('US Immigration Statistics (Ver 1.7.23).csv')
```

```python
# data cleaning

for col in df.columns[1:]:
    df[col] = df[col].apply(lambda x : x.replace(',','')).astype(float)
```

```python
df.info()
```

## Data Visualization

### Figure 1: Line Plot

```python
# Filter the DataFrame for the years 2011-2021
df = df[(df['Year'] >= 2011) & (df['Year'] <= 2021)]

# Extract columns
years = df['Year']
immigrants = df['Immigrants Obtaining Lawful Permanent Resident Status']
refugee_arrivals = df['Refugee Arrivals']
noncitizen_apprehensions = df['Noncitizen Apprehensions']
noncitizen_removals = df['Noncitizen Removals']
noncitizen_returns = df['Noncitizen Returns']

# Create the line plot
plt.figure(figsize=(12, 6))

# Add labels and title
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.title('US Immigration Statistics: Line Plot from 2011 to 2021', fontweight='bold')

# Plotting each category as lines with data points and adding labels
plt.plot(years, immigrants, marker='o', label='Immigrants Obtaining LPR Status')
plt.plot(years, refugee_arrivals, marker='o', label='Refugee Arrivals')
plt.plot(years, noncitizen_apprehensions, marker='o', label='Noncitizen Apprehensions')
plt.plot(years, noncitizen_removals, marker='o', label='Noncitizen Removals')
plt.plot(years, noncitizen_returns, marker='o', label='Noncitizen Returns')

# Add data labels as integers
for i, val in enumerate(immigrants):
    plt.text(years.iloc[i], val, str(int(val)), ha='center', va='bottom', color='blue', fontweight='bold')

for i, val in enumerate(refugee_arrivals):
    plt.text(years.iloc[i], val, str(int(val)), ha='center', va='bottom', color='orange', fontweight='bold')

for i, val in enumerate(noncitizen_apprehensions):
    plt.text(years.iloc[i], val, str(int(val)), ha='center', va='bottom', color='green', fontweight='bold')

for i, val in enumerate(noncitizen_removals):
    plt.text(years.iloc[i], val, str(int(val)), ha='center', va='bottom', color='red', fontweight='bold')

for i, val in enumerate(noncitizen_returns):
    plt.text(years.iloc[i], val, str(int(val)), ha='center', va='bottom', color='purple', fontweight='bold')

# Add legend
plt.legend()

# Set x-axis ticks to display all years
plt.xticks(years)

# Show the plot
plt.grid(True)
plt.show()
```

### Figure 2: Stem Plot

```python
# Create the stem graph
plt.figure(figsize=(12, 6))

# Add labels and title
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.title('US Immigration Statistics: Stem Plot from 2011 to 2021', fontweight='bold')

# Plotting each category as stems with markers
plt.stem(years, immigrants, linefmt='C0-', markerfmt='oC0', basefmt=" ", label='Immigrants Obtaining LPR Status')
plt.stem(years, refugee_arrivals, linefmt='C1-', markerfmt='sC1', basefmt=" ", label='Refugee Arrivals')
plt.stem(years, noncitizen_apprehensions, linefmt='C2-', markerfmt='^C2', basefmt=" ", label='Noncitizen Apprehensions')
plt.stem(years, noncitizen_removals, linefmt='C3-', markerfmt='dC3', basefmt=" ", label='Noncitizen Removals')
plt.stem(years, noncitizen_returns, linefmt='C4-', markerfmt='*C4', basefmt=" ", label='Noncitizen Returns')

# Add data labels as integers
for i, val in enumerate(immigrants):
    plt.text(years.iloc[i], val, str(int(val)), ha='left', va='bottom', color='blue', fontweight='bold')

for i, val in enumerate(refugee_arrivals):
    plt.text(years.iloc[i], val, str(int(val)), ha='left', va='bottom', color='orange', fontweight='bold')

for i, val in enumerate(noncitizen_apprehensions):
    plt.text(years.iloc[i], val, str(int(val)), ha='left', va='bottom', color='green', fontweight='bold')

for i, val in enumerate(noncitizen_removals):
    plt.text(years.iloc[i], val, str(int(val)), ha='left', va='bottom', color='red', fontweight='bold')

for i, val in enumerate(noncitizen_returns):
    plt.text(years.iloc[i], val, str(int(val)), ha='left', va='bottom', color='purple', fontweight='bold')
    
# Add legend
plt.legend()

# Set x-axis ticks to display all years
plt.xticks(years)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Extract columns
years = df['Year']
categories = {
    'Immigrants Obtaining Lawful Permanent Resident Status': {'color': 'C0', 'marker': 'o'},
    'Refugee Arrivals': {'color': 'C1', 'marker': 's'},
    'Noncitizen Apprehensions': {'color': 'C2', 'marker': '^'},
    'Noncitizen Removals': {'color': 'C3', 'marker': 'd'},
    'Noncitizen Returns': {'color': 'C4', 'marker': '*'}
}

# Create the stem graph
plt.figure(figsize=(12, 7))

# Add labels and title
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Percentage', fontweight='bold')  
plt.title('US Immigration Statistics: Percentage Stem Plot from 2011 to 2021', fontweight='bold')

# Calculate the total counts per year for normalization
total_counts_per_year = df[list(categories)].sum(axis=1)

# Plotting each category as stems with markers and normalize the values to percentages
for category, info in categories.items():
    normalized_values = (df[category] / total_counts_per_year) * 100  # Convert counts to percentages
    plt.stem(years, normalized_values, linefmt=info['color'], markerfmt=info['marker'] + info['color'], basefmt=" ", label=category, use_line_collection=True)

    # Add data labels as percentages
    for i, val in enumerate(normalized_values):
        plt.text(years.iloc[i], val, f"{val:.2f}%", ha='left', va='bottom', color=info['color'], fontweight='bold')

# Add legend
plt.legend()

# Set x-axis ticks to display all years
plt.xticks(years)

# Show the plot
plt.grid(True)
plt.show()
```

### Figure 3: Bar Plot

```python
# Height of a bar
bar_height = 0.18

# Positions of the bottom bar-boundaries
bar1 = np.arange(len(years))
bar2 = [x + bar_height for x in bar1]
bar3 = [x + bar_height for x in bar2]
bar4 = [x + bar_height for x in bar3]
bar5 = [x + bar_height for x in bar4]

# Create the horizontal bar plot
plt.figure(figsize=(15, 10))

# Plotting each category as a horizontal bar
# blue: #9bc8d9
# orange: #f1cca6
# green: #b4d19f
# red: #e59ea1
# brown: #e9d5d5
# purple: #cbb0e3
plt.barh(bar1, immigrants, height=bar_height, color='#9bc8d9', label='Immigrants Obtaining LPR Status')
plt.barh(bar2, refugee_arrivals, height=bar_height, color='#f1cca6', label='Refugee Arrivals')
plt.barh(bar3, noncitizen_apprehensions, height=bar_height, color='#b4d19f', label='Noncitizen Apprehensions')
plt.barh(bar4, noncitizen_removals, height=bar_height, color='#e59ea1', label='Noncitizen Removals')
plt.barh(bar5, noncitizen_returns, height=bar_height, color='#cbb0e3', label='Noncitizen Returns')

# Add labels and title
plt.ylabel('Year', fontsize=16, fontweight='bold')
plt.xlabel('Count', fontsize=16, fontweight='bold')
plt.title('US Immigration Statistics: Bar Plot from 2011 to 2021', fontsize=18, fontweight='bold')

# Add legend
plt.legend(fontsize=14)

# Set y-axis ticks to display all years with proper alignment
plt.yticks(bar2, years)

# Show the plot
plt.grid(True)
plt.show()
```
